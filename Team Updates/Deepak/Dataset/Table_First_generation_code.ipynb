```python
## Deriving the brand and category attributes
```

```python
def extract_brand(product_name):
    if pd.isna(product_name):
        return "Unknown"

    name = product_name.upper()

    if "HIK" in name or "EZVIZ" in name or "ECO" in name:
        return "Hikvision"
    elif name.startswith("CP") or "CP-" in name:
        return "CP Plus"
    elif "DAHUA" in name or name.startswith("DH-"):
        return "Dahua"
    elif "PRAMA" in name:
        
        return "Prama"
    elif "DAICHI" in name:
        return "Daichi"
    elif "IRANGE" in name:
        return "Irange"
    elif "BEETEL" in name:
        return "Beetel"
    else:
        return "Other"
```

```python
def extract_sub_category(product_name):
    if pd.isna(product_name):
        return "Unknown"

    name = product_name.upper()

    # --- Cameras & Core Devices ---
    if "VDP" in name or "DOOR PHONE" in name:
        return "Video Door Phone"
    elif "CAM" in name or "DOME" in name or "BULLET" in name:
        return "Camera"
    elif "NVR" in name or "DVR" in name:
        return "Recorder"
    elif "ACCESS CONTROL" in name or "PROXIMITY" in name:
        return "Access Control Unit"

    # --- Storage ---
    elif "HDD" in name or "HARD DISK" in name:
        return "Hard Drive"
    elif "SD" in name or "MICRO SD" in name:
        return "SD Card"

    # --- Locks ---
    elif "LOCK" in name:
        return "Door Lock"

    # --- Power ---
    elif "SMPS" in name:
        return "SMPS"
    elif "ADAPTER" in name or "POWER" in name:
        return "Power Adapter"
    elif "SOLAR" in name:
        return "Solar Panel"

    # --- Networking ---
    elif "POE" in name:
        return "POE"
    elif "SWITCH" in name:
        return "Network Switch"
    elif "CONNECTOR" in name or "BNC" in name or "RJ" in name:
        return "Connector"

    # --- Cabling (MORE SPECIFIC FIRST) ---
    elif "3+1" in name:
        return "3+1 Cable"
    elif "CAT 6" in name or "CAT6" in name:
        return "CAT 6 Cable"
    elif "HDMI" in name:
        return "HDMI Cable"
    elif "CABLE" in name or "WIRE" in name:
        return "Cable"

    # --- Mounting & Hardware ---
    elif "RACK" in name:
        return "Rack"
    elif "BRACKET" in name:
        return "Bracket"

    # --- Display & Input ---
    elif "MONITOR" in name:
        return "Monitor"
    elif "MOUSE" in name or "KEYBOARD" in name:
        return "Input Device"

    else:
        return "Other Accessories"
```

```python
def extract_category_from_sub(sub_category):
    camera_subs = [
        "Camera",
        "Video Door Phone",
        "Recorder",
        "Access Control Unit"
    ]

    if sub_category in camera_subs:
        return "Cameras"
    else:
        return "Accessories"
```

```python
df_copy["brand"] = df_copy["FIRM_PRODUCT"].apply(extract_brand)
```

```python
df_copy["sub_category"] = df_copy["FIRM_PRODUCT"].apply(extract_sub_category)
```

```python
df_copy["category"] = df_copy["sub_category"].apply(extract_category_from_sub)
```
